## numpy array vs python lists

**speed**

```python
#list 
a=[i for i in range(10000000)]
b=[i for i in range(10000000,20000000)]

c=[]
import time
start=time.time()
for i in range(len(a)):
    c.append(a[i]+b[i])
print(time.time()-start)
```

```python
# numpy
import numpy as np
a=np.arange(10000000)
b=np.arange(10000000,20000000)

start=time.time()

c=a+b
print(time.time()-start)
```

**mean-squared error**

##  Mean Squared Error (MSE) in NumPy

**Definition:**  
The **Mean Squared Error (MSE)** measures how far the predicted values are from the actual (true) values.  
It calculates the **average of the squared differences** between predictions and actual values.

---

###  Formula

\[
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_{true} - y_{pred})^2
\]

where  
- \( y_{true} \) = actual (real) values  
- \( y_{pred} \) = predicted values  
- \( n \) = number of samples  

---

###  In Simple Words:
- Find how **different** predictions are from actual values.  
- Square those differences (so negative errors dont cancel out).  
- Take the **average**  thats your MSE.  
- **Smaller MSE = better model.**

---
##  Step-by-Step (MSE calculation)

| y_true | y_pred | Error (y_true - y_pred) | Error |
|--------:|-------:|------------------------:|-------:|
| 3      | 2.5    | 0.5                     | 0.25   |
| -0.5   | 0.0    | -0.5                    | 0.25   |
| 2      | 2      | 0                       | 0      |
| 7      | 8      | -1                      | 1      |

\[
\text{MSE} = \frac{0.25 + 0.25 + 0 + 1}{4} = 0.375
\]

Or inline: `MSE = (0.25 + 0.25 + 0 + 1) / 4 = 0.375`

---

###  Uses
- **Regression models** (to measure prediction accuracy)  
- **Machine learning evaluation**  
- **Optimization** (loss to minimize during training)

---

###  Summary Table

| Term        | Meaning                       |
|-------------|-------------------------------|
| MSE         | Mean of squared errors        |
| Ideal Value | 0 (perfect prediction)        |
| Larger MSE  | Worse model performance       |



```python
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
```

## working with missing values

##  Working with Missing Values in NumPy (`np.isnan()`)

**Definition:**  
In NumPy, missing or undefined numeric data is often represented as **NaN** (Not a Number).  
To find or handle such missing values, we use the function **`np.isnan()`**.

---

###  What is `NaN`?
- **NaN** stands for **Not a Number**.  
- It represents missing, invalid, or undefined numeric data.  
- Example: result of `0/0` or invalid operations.

---
####  Summary Table

| Function | Description |
|-----------|--------------|
| `np.nan` | Represents missing value |
| `np.isnan(x)` | Returns True where `x` is NaN |
| `np.sum(np.isnan(x))` | Counts missing values |
| `x[~np.isnan(x)]` | Removes NaN values |
| `x[np.isnan(x)] = value` | Replaces NaN with given value |


```python
a=np.array([1,2,3,4,np.nan,6])
a
```

```python
np.isnan(a)
```

```python
a[~np.isnan(a)]
```

##  Plotting a Graph using NumPy

**Definition:**  
NumPy is mainly used for **numerical calculations**, but we can use it to **create data** that we then plot using **Matplotlib**.

---
###  Common NumPy Functions for Plotting Data

| Function | Description |
|-----------|--------------|
| `np.linspace(a, b, n)` | Creates evenly spaced numbers |
| `np.arange(start, stop, step)` | Creates range of values |
| `np.sin(x)` | Sine values |
| `np.cos(x)` | Cosine values |
| `np.exp(x)` | Exponential function |
| `np.log(x)` | Natural log values |


**plotting  a 2D plot**

```python
# x=y   
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y=x
plt.plot(x,y)
```

```python
# y=x**2
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
```

```python
# y=sin(x)
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
# y=xlog(x)
x=np.linspace(-10,10,100)
y=x*np.log(x)
plt.plot(x,y)
```

```python
# sigmoid
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
```
